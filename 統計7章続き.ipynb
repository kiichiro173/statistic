### 二次元正規分布の構成

標準正規分布N(0,1)に従う独立な二つの確率変数$X_1$,$X_2$があるときa,b,c,dを定数として確率変数の変換を行った時の期待値、分散、共分散、相関係数、同時確率密度関数を求めてみる。<br>
$Y_1 = aX_1 + bX_2$,$Y_2 = cX_1 + dX_2$と置く。


(a)$Y_1 , Y_2$の期待値$μ_1$ , $μ_2$<br>
(b)分散$σ_1^2$,$σ^2_2$、共分散$σ_{12}$、相関係数ρ<br>
(c)$Y_1$,$Y_2$の同時確率密度関数$g(y_1,y_2)$<br>

#### 期待値
今、$X_1$,$X_2$ともに標準正規分布に従っているため、$E(X_1)$=$E(X_1)$=0より、<br>
$E(Y_1)$ = $E(Y_2)$ = 0となる<br>
$μ_1$ = $μ_2$ = 0


#### 分散
分散の性質より,<br>
$V(Y_1)$ = $V(aX_1 + bX_2)$ = $a^2V(X_1) + b^2V(X_2)$<br>
また標準正規分布に従っているので$V(X_1)$ = $V(X_2)$ = 1<br>
よって<br>
$V(Y_1)$ = $a^2 + b^2=σ_1^2$<br>
同様に<br>
$V(Y_2)$ = $c^2 + d^2=σ_2^2$


#### 共分散
$Cov(Y_1 , Y_2)$ = $E\{(Y_1 - μ_1)(Y_2 - μ_2)\}$より<br>
$μ_1$ = $μ_2$ = 0なので<br>
$Cov(Y_1 , Y_2)$ = $E\{(Y_1)(Y_2)\}$=$E\{(aX_1 + bX_2)(cX_1 + dX_2)\}$よって<br>
$Cov(Y_1 , Y_2)$ = ac + bd



#### 相関係数

相関係数ρ=$Cov(Y_1 , Y_2)$/$\sqrt{V(Y_1)}\sqrt{V(Y_2)}$より<br>
ρ=$σ_{12}/σ_1σ_2$

####  同時確率密度関数

・一次元の正規分布<br>
$f(x) = \frac{1}{\sqrt{2π}σ}・exp\{-\frac{(x-μ)^2}{2σ^2}\}$(平均：μ、分散：$σ^2$)<br>
<br>

・標準正規分布関数(平均μ：0 , 分散$σ^2$：1)
<br>
$f(x) = \frac{1}{\sqrt{2π}}・exp\{-\frac{x^2}{2}\}$

<br>
・まず、$X_1 , X_2$の同時確率密度関数を考える。<br>
$X_1 , X_2$はどちらも標準正規分布に従っているため、それらの同時確率分布は確率密度関数の積となる。<br>
よって、<br>
$f(x_1 , x_2) = \frac{1}{2π}・exp\{-\frac{(x_1^2+x_2^2)}{2}\}$…①

<br>
<br>
・次に$Y_1 , Y_2$の同時確率密度関数$g(y_1 , y_2)$を考える。<br>
$y_1 = ax_1 + bx_2$ ,  $y_2 = cx_1 + dx_2$より、<br>
$x_1 = a^{'}y_1 + b^{'}y_2$ ,  $x_2 = c^{'}y_1 + d^{'}y_2$<br>
(ただし、$a^{'} = d/D$   ,    $b^{'} = -b/D$   ,  $c^{'} = -c/D$    ,$d^{'} = a/D$(D = ad - bc))<br>
つまり、<br>
$f(x_1 , x_2) = f(a^{'}y_1 + b^{'}y_2 , c^{'}y_1 + d^{'}y_2)$<br>
<br>
・４点$(y_1 , y_2)$ = (0 ,0), (0,1) , (1,0) , (1,1)の作る面積１の単位正方形が$x$の空間に移された平行四辺形の面積を調べると$|a^{'}d^{'} - b^{'}c^{'}|$となる。また、$|a^{'}d^{'} - b^{'}c^{'}|=1/|D|$である。<br>
これより同時確率密度関数は、①に代入して解くと<br>
$g(y_1 , y_2) = f(a^{'}y_1 + b^{'}y_2 , c^{'}y_1 + d^{'}y_2)・(1/|D|)$となる。<br>
<br>
・次に$f$の中を計算すると、$x_1^2 + x_2^2$ = $(a^{'}y_1 + b^{'}y_2 )^2 + (c^{'}y_1 + d^{'}y_2)^2$ = $Ay_1^2 + Cy_1y_2 + By_2^2$<br>
(ただし、$A = 1/σ_1^2(1-ρ^2)$,$B = 1/σ_2^2(1-ρ^2)$,$C = -2ρ/σ_1σ_2(1-ρ^2)$である。)<br>
<br>
ここで、恒等式$(ac+bd)^2 + (ad-bc)^2 = (a^2 + b^2)(c^2 + d^2)→(ad-bc)^2=(a^2 + b^2)(c^2 + d^2) - (ac+bd)^2より $<br>
$D^2=σ_1^2・σ_2^2-σ_{12}^2=σ_1^2・σ_2^2(1-ρ)$となる。<br>
<br>
以上より、$(Y_1 , Y_2)$の同時確率密度関数は、<br>
$g(y_1 , y_2) = f(a^{'}y_1 + b^{'}y_2 , c^{'}y_1 + d^{'}y_2)・(1/|D|)$=$\frac{1}{2πσ_1σ_2\sqrt{(1-ρ^2)}}・exp\{-\frac{1}{2(1-ρ^2)}(\frac{y_1^2}{σ_1^2}-\frac{2ρy_1y_2}{σ_1σ_2}+\frac{y_2^2}{σ_2^2})\}$→二次元正規分布

### 分散の加法性

・期待値の加法性<br>　
期待値は常に、<br>
E(X+Y) = E(X) + E(Y)が成り立つ。<br>
<br>
<br>
・分散の加法性<br>
XとYが独立である時、<br>
V(X±Y) = V(X) + V(Y)<br>
<br>
独立でないとき、<br>
$V(X+Y) = V(X) + V(Y) + 2Cov(X,Y) = V(X) + V(Y) + 2ρ_{XY}D(X)D(Y)$<br>
<br>
・ｎ個の場合<br>
$E(X_1 , X_2 , …X_n) = E(X_1) + E(X_2) + …E(X_n)$<br>
<br>
$X_1 , X_2 , …X_n$が独立である時、<br>
$V(X_1 , X_2 , …X_n) = V(X_1) + V(X_2) + …V(X_n)$<br>

<br>

#### 和の確率分布　

離散型の確率変数X,Yが独立であるとし、その確率分布をg(x),h(y)とする。和X+Yの確率分布k(z)は確率P(X+Y = z)を考えると得られる。$X+Y=z$となるのは、$X=x,Y=z-x$の形のx,z-xの加えてｚになるすべての組み合わせであるので、<br>
$k(z) = \sum_{x}g(x)h(z-x)$となる。<br>
関数g,hからｋを作るすうがく操作をg,hの畳み込みといいk=g*hとかく。<br>
<br>
g,hが密度関数の時も同様で畳み込みは積分<br>
$k(z) =   \int_{-∞}^{∞} g(x)h(z-x) dx$となる。ｐ150～


### 大数の法則
ベルヌーイ試行…一回の実験で２種類の事象（表/裏、成功/失敗など）のいずれかが生じしかもその様な事象が生起する確率が常に一定であるような試行のこと<br>
<br>
例：i回目のコイン投げで表が出た場合１、裏が出た場合0をとる確率変数$x_i$を考える。10回のコイン投げで表の出た回数$ｒ$は<br>
$r =x_1 +x_2 + …x_10$ <br>
である。<br>
<br>
表の出た回数の割合→$\hat{p}=r/10$<br>
一般にｎをコイン投げの回数とする時、r/nは相対頻度である。<br>
<br>
例の場合、rは確率変数で、n=10,p=0.5の二項分布$Bi(10 , 0.5)$すなわち、<br>
<br>
$f(10)={}_{10} C _x(1/2)^{10}$$(x = 1,2…10)$に従う。<br>
期待値:$E(r)…np=5$,分散:$V(r) = np(1-p)=5$<br>
<br>
また、割合$r/n$の期待値：$E(r/n)=p=0.5$、分散：$V(r/n)=p(1-p)/n=0.025$である。<br>
これは表が出る確率は0.5周辺の発生確率が高いことを表している。<br>
p157~$n=10$からｎを増やしたときのr/nの期待値とその周辺の発生する確率に関して調べている。<br>
<br>
大数の法則とは簡単に言うと、試行回数を増やせば増やすほど真の平均に近づく、真の平均を知ることが出来る。→ということはサンプル数を増やせば増やすほど母集団の特性を正確に知ることが出来る。

### コンピュータシミュレーション
p=0.4のベルヌーイ試行を2万回おこなってみる。100回おきに区切り1回目からの観測された成功率を計算し0.4に近づいていくのかを見てみる。

```
import numpy as np
def Ber(n):
  coin = [0,1]
  prob = [6/10 , 4/10]
  trials = [np.sum(np.random.choice(coin , i , p = prob)) / i for i in range(100 , n + 1 , 100)]
  return trials
```

```
#回数を重ねるごとに0.4に収束していっているのが分かる
import matplotlib.pyplot as plt
prob_list = Ber(20000)
left = prob_list
height = [i for i in range(100 , 20001 , 100)]
plt.plot(height , left)
```
