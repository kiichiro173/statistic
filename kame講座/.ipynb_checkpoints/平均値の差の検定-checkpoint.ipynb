## 平均値の差の検定(t検定)
t分布を使用するため、一般的にt検定と呼ばれる。

平均値の差の検定には大きく2種類存在する。
・対応がない
・対応がある

対応がないとは？
→実験Aと実験Bの被験者は全くの別人である。みたいな感じ
→この場合、実験Aと実験Bの母集団は別になってくる。
→よって調べるのは母集団Aと母集団Bの平均に差はあるの？みたいなことを考える必要がある。

対応があるとは？
→実験Aと実験Bの被験者が同じ。要は薬を与えたあとにその人は変化があったかなかったかを調べる時に使用する。
→この場合、母集団は同じである。よって、個々の差の平均が０なのかそうじゃないのかを検定していく必要がある。
→ここの差の平均が０である場合は、薬が意味なかったよねーとなる。なので個々の差が0でないと言えるかどうかが焦点になってくる。
