### 確率変数と確率分布

確率変数…各値に関してそれぞれの確率が与えられている変数のこと
<br>
<br>
例:

さいころを投げて出る目は｛1, 2, 3, 4, 5, 6｝のいずれかであり、それぞれの目が出る確率は$\frac{1}{6}$であると仮定すると、さいころを投げて出る目は確率変数であると言えます。<br>
<br>
<br>

確率分布…確率変数Xの値ごとの確率の分布

### 離散型の確率分布

一般に可算集合{$x_1$,$x_2$,…}の中の値をとる確率変数Xは離散型といわれ、それぞれの値の確率<br>
$P(X = x_k) = f(x_k)$  (k=1,2,…)<br>
をXの確率分布という。<br>
また、fは<br>
$f(x_k)≧0(k = 1,2,…)$かつ$\sum_{k = 1}^{∞}f(x_k) = 1$<br>
の条件を満たす。この$f$を離散型の確率分布という。

### 連続型の確率変数と確率分布
確率変数Xのとる値が関数$f(x)$によって<br>
$P(a≦X≦b) = \int_a^b f(x) dx$<br>
と表される場合、Xは連続型の確率分布を持つという。<br><br>
ただし<br>
すべてのxに対し<br>　
$f(x)≧0$かつ$ \int_{-∞}^∞ f(x) dx = 1$<br>
であるものとし、関数$f(x)$をXの確率密度関数という。<br>
<br>
またa = bと置けば<br>
$P(a≦X≦b)$→$P(X=a)$ = 0 <br>
となるので連続型の確率分布は一点の確率は0となる。
<br>
<br>
確率変数Xが確率分布f(x)を持つときXはf(x)に従うという。<br>
p91に連続型確率分布の例として、指数分布・一様分布が紹介されている。


### 累積分布関数

累積分布関数…確率変数Xがある値ｘ以下（X≦ｘ）の値となる確率」を表す関数。累積分布関数はF(x)で表されることが多い。
<br>
$F(x) = P(X≦x)$<br>
<br>
離散型の累積分布関数…$\sum_{u≦x}f(x_u)$<br>
連続型の累積分布関数…$F(x) = \int_{-∞}^{x} f(u) du$
<br>
確率密度関数は累積密度関数を微分することで求めることができる。<br>
$f(x) = F^{'}(x)$<br>
<br>
また、累積分布関数は常に３つの性質を持つ<br>
<br>
1.広義単調増加<br>
$x_1 < x_2$ならば、$F(x_1) ≦ F(x_2)$<br>
<br>
2.範囲<br>
$x→∞の時　F(x)→1$<br>
$x→-∞の時　F(x)→0$<br>
<br>
3.右連続<br>
$各点xでε（イプシロン）↓0の時F(x + ε) →　F(x)$

↓0：0に近づける
関数の右連続性：x が点 a に右から近づいたとき，関数 f(x) がとぎれることなく a までたどりつけるとき f(x) は x=a で右連続と言います。



### モード・メディアン

モード…最頻値のこと。確率密度関数$f(x)$を最大にする$x=x_0$であって、この値が最も出る確率が高いという意味で確率分布を代表する。<br>
<br>
メディアン…平均値のこと。$P(X≦x_m) = 1/2$であり真ん中の値。期待値でもある。


### 期待値

期待値…1回の試行で得られる値の平均値のことで、得られうるすべての値とそれが起こる確率の積を足し合わせたもの<br>
<br>
例.さいころを1回投げるとき、出る目の期待値<br>
さいころの目をXとすると、<br>
$E(X) = 1・(1/6) + 2・(1/6) +…+6・(1/6) = 3.5$
<br>したがって、さいころを何回か投げると1回当たりの出る目の平均は3.5になると期待できます。
<br>
<br>
離散型の期待値…$\sum_{x}x・f(x)$<br>
連続型の期待値…$\int_{-∞}^{∞}x・f(x)dx$<br>
<br>
期待値は無限和・積分であるため存在しないこともある。（発散するとき）<br>
無限和が存在しない…https://yama-taku.science/mathematics/series/examples-of-series/<br>
変数Xの関数Φ(X)についても同様に期待値を定義する。（Φファイ）<br>
離散型の期待値…$\sum_{x}Φ(x)・f(x)$<br>
連続型の期待値…$\int_{-∞}^{∞}Φ(x)・f(x)dx$<br>
p95に期待値の例<br>
<br>
・期待値の性質<br>
・E(c) = c <br> 
・E(X + c) = E(X)  + c <br> 
・E(cX) = c・E(X)<br> 
・E(X + Y) = E(X) + E(Y) <br> 


### 分散と標準偏差
<br>
分散…期待値μ＝$E(X)$であるとしたとき、分散V(X)は$E({(X-μ)}^2)$と定義する。<br>
定義より<br>
<br>
離散型の分散…$\sum_{x}(x-μ)^2・f(x)$<br>
連続型の分散…$\int_{-∞}^{∞}(x-μ)^2・f(x)dx$<br>
<br>
V(X) = $E(X^2) - (E(X))^2$
<br>
<br>
標準偏差…標準偏差は分散の平方根で求めることができる。$\sqrt{V(X)}$
<br>
分散の性質<br>
$V(c) = 0 $<br>
$V(X + c) = V(X)$<br>
$V(cX) =c^2・V(X) $<br>

### モーメント

モーメント…積率ともいわれ、分布の平均値やバラツキ、ひずみや尖り度を数値化するものである。（https://www.aandt.co.jp/jpn/qc/basic/moment.htm）

Xの(原点周りの)r次のモーメント…$μ_r = E(X^r)$（r=1の時→期待値）<br>
Xの期待値周りのr次のモーメント…$μ^{'}_r = E(X-μ)^r$(ただし$μ=E(X)$)（分散）<br>
標準化モーメント…$α_r = E((X-μ)/σ)^r$（尖度や歪度）<br>
<br>
積率母関数（モーメント母関数）…$M_x(t) = E(e^{tx})$と定義する。$e^{tx}$の期待値を考えている(Xは確率変数・f(x)はｘの確率)<br>
離散型‥$M_X(t) = \sum_{x}e^{tx}・f(x)$<br>
連続型‥$M_X(t) = \int_{-∞}^{∞}e^{tx}・f(x)dx$<br>
<br>
この積率母関数は何がいいのか？？<br>
→積率母関数を微分することにより、確率変数Xのr乗の期待値を求めることができる。（各次数のモーメントが分かる）積率母関数を使用すると簡単に期待値や分散などを求めることができるってこと。<br>
例<br>
$M_X^{'}(0) = μ_1$,$M_X^{''}(0) = μ_2$,…$M_X^{r}(0) = μ_r$  ($μ_r = E(X^r)$)<br>

### チェビシェフの不等式
チェビシェフの不等式…確率分布はわかっていないけど、わかっている期待値と分散から確率を求めるときに役に立つ。どのような確率変数Xもその期待値（平均）E(X)の周辺に確率が集中しており、期待値から離れれば離れるほど確率は低くなるがその程度は分散$σ^2 = V(X)$による。標準偏差$σ=D(X)$を単位にとるとき、一般に次の不等式が成り立つ。<br>
$P(|X-μ|≧kσ)≦1/k^2 , (ただしk>0、μ=E(X),σ^2=V(X))$
